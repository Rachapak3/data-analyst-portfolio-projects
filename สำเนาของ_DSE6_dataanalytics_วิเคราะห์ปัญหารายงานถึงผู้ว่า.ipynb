```python
data['น้ำ'] = (data['comment'].str.find('น้ำ')!=-1).astype('int')
```

```python
pdic = {}
pdic['น้ำท่วม'] = ['น้ำท่วม','น้ำขัง','ระบายน้ำ','ฝนตก']
pdic['มืด'] = ['มืด','สว่าง','เปลี่ยว']
pdic['จราจร'] = ['รถติด','จราจร']
pdic['ทางข้าม'] = ['ม้าลาย','ข้ามถนน']
pdic['กลิ่น'] = ['สกปรก','ขยะ','เหม็น','กลิ่น','น้ำเสีย']
pdic['สัตว์'] = ['แมว','สุนัข','จรจัด','สัตว์']
pdic['ทางเดิน'] = ['ทางเดิน','ทางเท้า','ฟุตบาท','กีดขวาง','ทางข้าม','จอดรถ']
pdic['ถนน'] = ['หลุม','ป้ายบอกทาง','ถนนขรุขระ']
pdic['ไม่เป็นระเบียบ'] = ['สายไฟ','ไม่เป็นระเบียบ']
pdic['ต้นไม้'] = ['ต้นไม้']
pdic['ไม่ปลอดภัย'] = ['ไม่ปลอดภัย','อันตราย']
```

```python
import numpy as np
def createfeat(data):
  ndata = data.copy()
  for k in pdic.keys():
    wList = pdic[k]
    dList = []
    for w in wList:
      dList.append(data['comment'].str.find(w)!=-1)
    dList = (np.vstack(dList).sum(0)>0).astype('int')
    ndata[k] = dList
  return ndata
```

```python
data2 = createfeat(data)
```

```python
from wordcloud import WordCloud
from collections import Counter
def createfig(prob):
  path = '/content/THSarabunNew.ttf'
  datd = data2[data2[prob]==1]
  textList = datd['comment'].values
  senList = []
  for t in textList:
    senList = senList + pn.word_tokenize(t)
  word_cloud_dict=Counter(senList)
  fname = prob+'.jpg'
  genwordcloud(word_cloud_dict,regexp,fname)
```

```python
def genwordcloud(word_cloud_dict,regexp,fname):
  wordcloud = WordCloud(
                        font_path=path,
                        relative_scaling = 0.3,
                        min_font_size=1,
                        background_color="white",
                        width=1024,
                        height=768,
                        max_words=2000,
                        colormap='plasma',
                        scale=3,
                        font_step=4,
                        regexp=regexp,
                      #   contour_width=3,
                      #   contour_color='steelblue',
                        collocations=False,
                        margin=2
                        ).generate_from_frequencies(word_cloud_dict)
  import pylab as plt
  plt.rcParams.update({'font.size': 22})
  fig, ax = plt.subplots(1, 1, figsize=(16, 12))
  ax.imshow(wordcloud, interpolation='bilinear')
  ax.set_title(fname.split('.')[0],)
  ax.axis("off")
  fig.show()
  fig.savefig(fname)
```

```python
pList = list(pdic.keys())
for p in pList:
  print(p)
  createfig(p)
```
